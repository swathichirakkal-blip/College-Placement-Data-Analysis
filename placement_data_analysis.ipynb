## Data Cleaning

```python

print("\nNull values in each coloumn:")
print(data.isnull().sum())
data=data.dropna()
dat=data.drop_duplicates()
print("\nShape after cleaning:",data.shape)
```

## Exploratory Data Analysis

```python
placement_year=data.groupby("year")["college name"].count()

placement_year
```

## Visualization

```python
import matplotlib.pyplot as plt
placement_year=data.groupby("year")["college name"].count()

placement_year.plot(kind='line',marker='o')

plt.title("Students placed trend by year")
plt.xlabel("year")
plt.ylabel("Number of students placed")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("placement analysis.xlsx")
data['name of company'].value_counts().head(6).plot.pie(autopct='%1.1f%%')
plt.title("Top Hiring Companies")
plt.show()
```

```python
avg_salary = data.groupby("college name")["Salary"].mean()

top10 = avg_salary.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top10.sort_values().plot(kind='barh')

plt.title("Top 10 colleges by Average Placement Salary", fontsize=14, fontweight='bold')
plt.xlabel("Average Salary (LPA)")
plt.ylabel("college Name")


plt.show()
```

## Key Insights

-Placement performance has shown consistent improvement over years
-IT service companies dominate campus recruitment,acting as primary hiring partners
-Salary distribution varies significantly across colleges ,indicating competitive positioning among institutions
-The sharp growth observed in 2018 suggests increased industry demand and improved campus placement readiness
